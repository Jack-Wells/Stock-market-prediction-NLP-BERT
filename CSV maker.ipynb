```python
import re
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
file = open('stocks_name.txt','r')

Name = []
Codes = []
Type = []

for line in file:
    line = re.sub('\n', '', line)
    line = line.split('\t')
    Name.append(re.sub(' ', '-',line[0].lower()))
    Type.append(line[2])
    if line[1][2:4] == '.A':
        Codes.append('BT-A.L')
    elif line[1][2] == '.':
        Codes.append(line[1] + 'L')
    else:
        Codes.append(line[1] + '.L')
print(Codes)
```

```python
for code in Codes:
    data = yf.download(code,'2014-01-01','2020-04-01')
    data.to_csv('csv/'+ code + '.csv')
```
