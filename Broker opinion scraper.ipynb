## Web Scraper for Trader Opinion Mining

```python
import requests
from   bs4 import BeautifulSoup
import re
```

```python
file = open('stocks_name.txt','r')

Name = []
Codes = []
Type = []

for line in file:
    line = re.sub('\n', '', line)
    line = line.split('\t')
    Name.append(line[0])
    Type.append(line[2])
    Codes.append(line[1])
print(Codes[:])
print(Codes[80])
```

```python
file = open('Brokers4.csv','w',encoding='UTF-8')

All = []
Codes = ['RSA']
for code in Codes:
    
    previous = []
    data = []
    
    for k in range(1,11):
        r = requests.get('https://www.lse.co.uk/ShareBrokerTips.asp?shareprice='+ code + '&share=' + code + '&page=' + str(k),
                         headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(r.content, 'html.parser')
        kap = soup.find_all('td')

        Break = False
        In = []
        line = []
        for i in range(len(kap)):
            line.append(kap[i].text)
        for j in range(int(len(kap)/6)):
            im = []
            for i in range(6):
                im.append(line[j*6+i])
            In.append(im)
            if previous == In:
                print(k)
                Break = True
                break
            data.append(im)
        print(data)
        if Break == True:
            data = data[:-len(In)+1]
            break
        previous = In
        #data.append(im)
        
        
    All.append(data)
    
    file.write(str(data))
    file.write('\n')
print(len(All))
```
